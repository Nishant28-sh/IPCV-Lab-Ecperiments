```python
img =cv2.imread('/content/image.jpeg', 0 ) #grayscale0
```

```python
kernel= np.ones((5,5), np.float32)/25
img_blur= cv2.filter2D(img, -1, kernel)
```

```python
cv2.imwrite("1.png", np.abs(img_blur))
```

```python
noise= 20*np.sin(np.linspace(0,50,img.shape[1]))
```

```python
noise = noise.reshape(1,-1)
```

```python
noisy = img_blur+noise
```

```python
cv2.imwrite("2.png", np.abs(noisy))
```

```python
noisy = np.clip(noisy, 0, 255).astype(np.uint8)
```

```python
f= np.fft.fft2(noisy)
```

```python
fshift= np.fft.fftshift(f)
```

```python
h=np.fft.fft2(kernel,s=noisy.shape)
h_shift=np.fft.fftshift(h)
```

```python
epsilon= 1e-5
```

```python
h_shift=np.where(h_shift ==0,epsilon, h_shift)
```

```python
g=fshift/h_shift
```

```python
g_ishift=np.fft.ifftshift(g)
```

```python
g_iift=np.fft.ifft2(g_ishift)
```

```python
inv_img= np.abs(g_iift)
```

```python
cv2.imwrite("3.png", np.abs(inv_img))
```

```python
weiner_img= wiener(noisy)
```

```python
cv2.imwrite("4.png", np.abs(weiner_img))
```
