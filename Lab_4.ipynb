```python
img =cv2.imread('/content/eagle_img.webp', 0 ) #grayscale0
```

```python
f= np.fft.fft2(img)
```

```python
fshift= np.fft.fftshift(f) # low mid mai aa gai
```

```python
row, col= img.shape
```

```python
crow, ccol= row//2,col//2
```

```python
mask_lp=np.zeros((row,col),np.uint8)
```

```python
mask_lp[crow-30:crow+30,ccol-30:ccol+30]=1
```

```python
lp=np.fft.ifft2(np.fft.ifftshift(fshift*mask_lp))
```

```python
cv2.imwrite("lpf_eagle.png", np.abs(lp))
```

```python
mask_hp=1-mask_lp
```

```python
mask_hp[crow-30:crow+30,ccol-30:ccol+30]=0
```

```python
hp=np.fft.ifft2(np.fft.ifftshift(fshift*mask_hp))
```

```python
cv2.imwrite("hpf_eagle.png", np.abs(hp))
```
