```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img= cv2.imread('/content/image.png')
```

```python
rows, cols = img.shape[:2]
```

```python
M_trans= np.float32([[1,0,50],[0,1,50]])
```

```python
translated=cv2.warpAffine(img,M_trans,(cols,rows))
```

```python
cv2.imwrite('translated.png',translated)
```

```python
scaled=cv2.resize(img,None,fx=0.5,fy=0.5)
```

```python
cv2.imwrite('translatedresized.png',scaled)
```

```python
M_rot=cv2.getRotationMatrix2D((cols/3,rows/3),60,1)
```

```python
rotated=cv2.warpAffine(img,M_rot,(cols,rows))
```

```python
cv2.imwrite('rotated.png',rotated)
```
